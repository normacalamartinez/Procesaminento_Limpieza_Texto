## SoluciÃ³n de taller de Limpieza y procesamiento de texto.

### Punto 1
a. Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con with open

```python
# ğŸ“ Leer plano.

with open("UNA_SOLITARIA_VOZ_HUMANA.txt",encoding="utf-8") as archivo: 
    data = archivo.read()
print(data)
```

b. Convertir a minÃºsculas

```python
# Resultado tokenizado
texto_lower = data.lower() # MinÃºscula
print(texto_lower)
```

c. Conservar sÃ³lo palabras

```python
import re 
texto_letras = re.sub(r"[\W\d_]+", " ", texto_lower) # Dejar sÃ³lo letras
texto_sin_espacio = texto_letras.strip() # Quitar espacios del principio.
texto_sin_espacio
```

d. Tokenizar

```python
texto_tokenizado = texto_sin_espacio.split() # TokenizaciÃ³n
texto_tokenizado
```

## Punto 2:

a. Quitar palabras vacias

```python
# Quitar palabras vacias
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
texto_sin_vacios = [palabra for palabra in texto_tokenizado 
         if palabra not in stopwords_sp]
texto_sin_vacios
```

b. Â¿CuÃ¡les son las 10 palabras no vacias mÃ¡s usadas?

```python
from collections import Counter

# MÃ¡s comunes
Counter(texto_sin_vacios).most_common(10)
```

## Punto 3:

a. Stemming del documento

```python
from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")
```

```python
print("Cantidad de palabras raiz del texto\n")

texto_raiz = []

for palabra in texto_sin_vacios:
    texto_raiz.append(spanishStemmer.stem(palabra))

texto_raiz 
          
```

b. Â¿CuÃ¡les son las 10 raÃ­ces mÃ¡s usadas?

```python
Counter(texto_raiz).most_common(10)
```

### Punto 4

a. Contar cuÃ¡ntas ocurrencias hay por cada parte de la oraciÃ³n.

```python
#pip install spacy
#python -m spacy download es_core_news_sm
```

```python
import es_core_news_sm

nlp = es_core_news_sm.load()
doc = nlp( ' '.join(texto_sin_vacios))

ocurrencias = []

for w in doc:
    ocurrencias.append(w.pos_)    
    
Counter(ocurrencias).most_common(10)
```
